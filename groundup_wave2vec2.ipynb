```python
# Import necessary library

# For managing audio file
import librosa

#Importing Pytorch
import torch

#Importing Wav2Vec
from transformers import Wav2Vec2ForCTC, Wav2Vec2Tokenizer
```

```python
# Reading taken audio clip

file_name = "malinkundangtest2.wav"
import IPython.display as display
display.Audio(file_name, autoplay=True)
# Loading the audio file

audio, rate = librosa.load(file_name, sr = 16000)
```

```python
display.Audio(file_name, autoplay=True)
```

```python
from transformers import Wav2Vec2ForCTC, Wav2Vec2Processor
model_name = "indonesian-nlp/wav2vec2-large-xlsr-indonesian"
#model_name = "ayameRushia/wav2vec2-large-xlsr-indonesia-demo"
#model_name = "Galuh/wav2vec2-large-xlsr-indonesian-demo"

processor = Wav2Vec2Processor.from_pretrained(model_name)
model = Wav2Vec2ForCTC.from_pretrained(model_name)
```

```python
input_values = processor(audio, return_tensors = "pt", sampling_rate=16000).input_values
```

```python
logits = model(input_values).logits
```

```python
prediction = torch.argmax(logits, dim = -1)
```

```python
transcription = processor.batch_decode(prediction)[0]
```
